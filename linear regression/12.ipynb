```python
def calc_dp(sum_of_sq, size): #Bool
    dp = np.zeros((sum_of_sq + 1, size + 1), dtype=bool)
    dp[0][0] = True

    for i in range(1, size):
        for j in range(0, sum_of_sq):
            for k in range(0, 15):
                if j + k * k > sum_of_sq or not dp[j][i - 1]:
                    continue
                dp[j + k * k][i] = True
    return dp

def getlist(sum_of_sq, size, dp):
    cur = []
    for i in range(size):
        buf = [0]
        for j in range(1, 16):
            if sum_of_sq - j * j >= 0 and dp[sum_of_sq - j * j][size - 1]:
                buf.append(j)
        rand = np.random.randint(0, len(buf))
        cur.append(buf[rand])
        sum_of_sq = sum_of_sq - buf[rand] * buf[rand]
        size = size - 1

    return np.random.permutation(cur)

dp = calc_dp(2882, 100)
X = getlist(2882, 100, dp)
Y = getlist(2882, 100, dp)

print(sum(X * X), sum(Y * Y))
```

```python
rng = np.random.default_rng(42)
df = pd.DataFrame({"x" : X, "y" : Y})
formula = "y ~ x - 1"
model = ols(formula, df).fit()
print(model.params)
```

```python
import numpy as np
import solve
import random
```

```python
def test(i, j): #Decomposition test
    A = np.random.randn(i, j)
    B = solve.solve(A)
    assert np.allclose(A, B)

for i in range(10):
    test(random.randint(1, 7), random.randint(1, 7))
print("OK")
```

```python
def test_identity(): #Test identity matrix
    A = np.eye(4)
    B = solve.solve(A)
    assert np.allclose(A, B)

test_identity()
print("OK")
```

```python
def random_test_1(n):
    A = np.eye(4)
    B = np.zeros((4, n))
    A = np.concatenate( (A, B), axis=1 )
    B = solve.solve(A)
    assert np.allclose(A, B)

for i in range(2):
    random_test_1(i)
print("OK")
```

```python
def random_test_2(n):
    A = np.random.randn(4, 4)
    B = np.zeros((4, n))
    A = np.concatenate( (A, B), axis=1 )
    B = solve.solve(A)
    assert np.allclose(A, B)

for i in range(6):
    random_test_2(i)
print("OK")
```

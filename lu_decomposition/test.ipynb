```python
import numpy as np
import solve as sl
np.random.seed(42)
```

```python
def test_lu_decomposition(): #Decomposition test
    A = np.random.randn(5, 5)
    P, L, U = sl.lu(A)
    assert np.allclose(P @ A, L @ U)

for i in range(5):
    test_lu_decomposition()
print("OK")
```

```python
def test_lu_solve(): #Solution test
    A = np.random.randn(5, 5)
    b = np.random.randn(5)

    P, L, U = sl.lu(A)
    x = sl.solve(P, L, U, b)
    assert np.allclose(A @ x, b)

for i in range(10):
    test_lu_solve()
print("OK")
```

```python
def test_against_numpy(): #Compare with numpy
    A = np.random.randn(5, 5)
    b = np.random.randn(5)

    x_lu = sl.solve(*sl.lu(A), b)
    x_np = np.linalg.solve(A, b)

    assert np.allclose(x_lu, x_np)

for i in range(10):
    test_against_numpy()
print("OK")
```

```python
def test_identity(): #Test identity matrix
    A = np.eye(4)
    b = np.array([1, 2, 3, 4])

    x = sl.solve(*sl.lu(A), b)
    assert np.allclose(x, b)
test_identity()
print("OK")
```

```python
import pytest

def test_singular():
    A = np.array([[1, 2], [2, 4]])

    with pytest.raises(ValueError): #Test vr on linear dependency (0 on main diagonal)
        sl.lu(A)
test_singular()

print("OK")
```
